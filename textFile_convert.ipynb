```python
import time
start = time.time()
def create_test_file(filename, size_mb):
    line = "This is a sample line of text to convert.\n"
    with open(filename, "w", encoding="utf-8") as f:
        for _ in range(size_mb * 1024 * 1024 // len(line)):
            f.write(line)

create_test_file("200MB.txt", 200)
create_test_file("400MB.txt", 400)
create_test_file("600MB.txt", 600)
create_test_file("800MB.txt", 800)
create_test_file("1000MB.txt", 1000)


file_sizes = [200, 400, 600, 800, 1000]  # In MB
for size in file_sizes:
    input_file = f"{size}MB.txt"
    output_file = f"{size}MB_upper.txt"

    start = time.perf_counter()
    with open(input_file, 'r', encoding='utf-8') as f_in, open(output_file, 'w', encoding='utf-8') as f_out:
        for line in f_in:
            f_out.write(line.upper())
    end = time.perf_counter()

    print(f"{size}MB file processed in {end - start:.2f} seconds")

```

```python
import matplotlib.pyplot as plt

sizes = [200, 400, 600, 800, 1000]
python_times = [25, 40, 53, 75, 100]
c_times = [12, 20, 34, 45, 55]
cpp_times = [15, 25, 36, 50, 60]
java_times = [18, 30, 40, 55, 70]
r_times = [20, 35, 45, 60, 80]

plt.plot(sizes, c_times, label='C', marker='o')
plt.plot(sizes, cpp_times, label='C++', marker='o')
plt.plot(sizes, java_times, label='Java', marker='o')
plt.plot(sizes, r_times, label='R', marker='o')
plt.plot(sizes, python_times, label='Python', marker='o')

plt.title('Performance Comparison: Uppercase Conversion')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
